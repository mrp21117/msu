# Data Visualization

# Ploting

```python
#import required package
#conda install matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#x axis
x = [1,2,3,4,5]
#y axis
y = [20,5,15,10,30]
#plot points
plt.plot(x,y,color="green")

#lable
plt.xlabel("X-Axis")

#plt.ylabel("Y-Axis")
```

# Ploting Parameters

```python
#marker
#c -for line colour
#linesyle 
#ms = marker size
#mec = marker border
#mfc 

ypoints = np.array([20,5,15,10,30])
plt.plot(ypoints,marker='H',c="b",linestyle="dashdot",ms="10",mec="yellow",mfc="red")
plt.xlabel("Frequency")
plt.ylabel("Gain")
plt.title("Response")
plt.show()
```

# Bar Plot

```python
x=[2,4,6,8]
y=[10,20,30,40]
plt.bar(x,y,color="red")
```

```python
#scatter
x=[2,4,6,8]
y=[10,20,30,40]
plt.scatter(x,y,color="red")
```

```python
#Histogram
data = np.random.normal(170,10,250)
plt.hist(data)
plt.hist(data,bins=20)
plt.show()
```

```python
#bar plot
#ticks parameter
x=['CSE','MEC','ELE','EC \n Electronics']
y=[30,50,70,40]
plt.bar(x,y)
plt.xlabel("Dept")
plt.ylabel("Students")
plt.title("MSU Addmission")
plt.tick_params(axis="x",labelsize='10',labelrotation=45,labelcolor='blue')
plt.tick_params(axis="y",labelsize='10',labelrotation=90,labelcolor='blue')

plt.show()
```

# Legend

```python
y1 = [1,5,7,3,2]
y2 = [1,2,4,6,8]
plt.plot(y1)
plt.plot(y2)
plt.legend(["blue","orange"],loc='lower right')
```

# Text annotation

```python
#Draw Plot as figure
fig, ax=plt.subplots()
x = np.linspace(0,20,1000)
y = np.cos(x)
plt.plot(x,y)
ax.axis('equal')
ax.annotate("Maximum",xy=(2*np.pi,1),xytext=(10,4),arrowprops=dict(facecolor="black"))
ax.annotate("Minimum",xy=(5*np.pi,-1),xytext=(20,-5),arrowprops=dict(facecolor="black"))
plt.show()
```
